```python
# формируем группу риска (пользователи с более чем 1 заказом)
sizes = orders.groupby(['visitorId', 'group']).size().reset_index()
sizes.columns = ['visitorId', 'group', 'count']
sizes = sizes.query('count > 1').sort_values(by='visitorId')
sizes
```

**Формируем словарь "пользователь - группы, в которых он появлялся" и если одному пользователю принадлежит более 1 значения групп, добавляем его в список broken_users:**

```python
potentially_broken_users_list = list(zip(sizes['visitorId'], sizes['group']))
potentially_broken_users_dict = defaultdict(list)
for user, group in potentially_broken_users_list:
    potentially_broken_users_dict[user].append(group)

broken_users = [user for user in potentially_broken_users_dict if len(potentially_broken_users_dict[user]) > 1]
print('Пользователи, которые видели обе версии сайта: ', ', '.join(str(el) for el in broken_users))
```

Пять пользователей оказались в двух группах одновременно, они совершили достаточно много транзакций и принесли значимую суму выручки, поэтому нужно очистить наши таблицы от этих пользователей, для корретного А/В теста

```python
broken_orders = orders.query('visitorId in @broken_users')
print('Количество строк в датафрейме orders по этим пользователям: {}'.format(broken_orders.shape[0]))
broken_orders.head()
```

**Посмотрим, сколько заказов они сделали, на какую сумму и по каким группам:**

```python
broken_orders_grouped = (broken_orders
                        .groupby(['visitorId', 'group', 'date'])
                        .agg({'transactionId': 'nunique', 'revenue': 'sum'})
                   )
broken_orders_grouped
```

Они сделали в общей сложности 44 заказа и принесли компании немало выручки. Получается, что данные "испорчены" этими пользователями.

В связи с этим, создадим копии наших основных датафреймов **orders** и **visitors*** и в них уберем данные по этим пользователям из датафрейма **orders** и для корректности из таблицы **visitors** удалим по 1 broken-посетителю в соответствующих группах и в соответствии с датами, полученными выше. В конце выполнения задачи, рассчитаем стат. значимость для обоих наборов данных: с этими "неправильными" пользователями и без них, - чтобы проследить, влияют ли они на конечный результат.

Чистим датафреймы:

* orders:

```python
orders_filt = orders.query('visitorId not in @broken_users')
```

* visitors:


Сначала добавим к нашему кастомному датафрейму столбец count, где проставим значение -1, и возьмем только столбцы **group**, **date** и **count**:

```python
broken_orders_grouped = broken_orders_grouped.reset_index()
broken_orders_grouped['count'] = -1
broken_orders_grouped = broken_orders_grouped[['group', 'date', 'count']]
broken_orders_grouped.head()
```

Объединим наш начальный датайфрейм **visitors** с полученным выше:

```python
visitors_filt = visitors.merge(broken_orders_grouped, on=['date', 'group'], how='left').fillna(0)
visitors_filt.head()
```

Теперь добавим новый столбец, где построчно просуммируем значения столбцов **visitors** и **count**:

```python
visitors_filt['visitors_'] = visitors_filt['visitors'] + visitors_filt['count']
visitors.head()
```

Отбросим старый столбец **visitors** и вспомогательный столбец count, переименуем новый столбец **visitors_** в **visitors**:

```python
visitors_filt = visitors_filt.drop(['visitors', 'count'], axis=1)
visitors_filt.columns = ['date', 'group', 'visitors']
visitors_filt['visitors'] = visitors_filt['visitors'].astype(int)
visitors_filt.duplicated().any()
```

```python
visitors_filt = visitors_filt.drop_duplicates().reset_index()
```

```python
visitors_filt = visitors_filt.drop(columns="index")
```

<a id="2.2"></a>**2.2 Проверка корректности распределения на группы A/B**

[к содержанию](#0.1)

* Проверим размеры групп в выборке, есть ли разница между ними
* Проверим, корректно ли был рассчитан размер выборки для проведения теста

```python
print('Кол-во посетителей в группе А: {}\n'
      'Кол-во посетителей в группе B: {}'.format(visitors_filt[visitors_filt['group']=='A']\
                                                 ['visitors'].sum(), 
                                                 visitors_filt[visitors_filt['group']=='B']\
                                                 ['visitors'].sum()))
purchasesA = len(orders_filt[orders_filt['group']=='A'])
purchasesB = len(orders_filt[orders_filt['group']=='B'])
print('Кол-во покупок в группе A: {}'.format(purchasesA))
print('Кол-во покупок в группе B: {}'.format(purchasesB))
```

```python
plt.hist(visitors[visitors_filt['group']=='A']['visitors'], alpha = 0.8, label='A', bins = 30)
plt.hist(visitors[visitors_filt['group']=='B']['visitors'], alpha = 0.8, label='B', bins = 30)
plt.legend()
plt.title('Гистограммы кол-ва посетителей в день по группам A и B')
plt.show()
results = stats.mannwhitneyu(visitors_filt[visitors_filt['group']=='A']['visitors'],\
                             visitors_filt[visitors_filt['group']=='B']['visitors'])
alpha = 0.05
print('P-value:{}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Разница в количестве посетителей в группах A и B статистически значима\n')
else:
    print('Разница в количестве посетителей в группах A и B статистически НЕ значима\n')
```

```python
initial_conversion = len(orders_filt[orders_filt['group']=='A']) / visitors_filt[visitors_filt\
                                                            ['group']=='A']['visitors'].sum() 
resulting_conversion = len(orders_filt[orders_filt['group']=='B']) / visitors_filt[visitors_filt\
                                                                ['group']=='B']['visitors'].sum()
print('Начальная конверсия: {:.3f}'.format(initial_conversion))
print('Полученная после изменений конверсия: {:.3f}'.format(resulting_conversion))
print('Относительное изменение конверсии после изменений: {:.3f}'.format(abs(1-\
(initial_conversion/resulting_conversion))))
print('\nВводим начальную конверсию и полученное относительное изменение конверсии в калькулятор\n'
     'Сайт калькулятора: https://www.evanmiller.org/ab-testing/sample-size.html\n'
     'Получаем необходимое количество выборки 31868 для корректного проведения тестирования\n'
    'В нашем случае количество {} посещений\n'
      'Кол-во выборки для A/B тестирования корректно подобрано, ' 
      'статистически значимой разницы\nв количестве посещений м/у группами не обнаружено\n\n\
Ок, идём дальше.'.format(visitors_filt['visitors'].sum()))
```

<a id="2.3.1"></a>**2.3.1 График кумулятивной выручки по группам**

<a id="2.4.5"></a>**2.4.5 Статистическая значимость различий в конверсии между группами по «сырым» данным.**

[к содержанию](#0.1)

```python
print('Кол-во покупателей в группе A: {}'.format(len(ordersByUsersA)))
print('Кол-во покупок в группе А: {}'.format(ordersByUsersA['orders'].sum()))
print('Кол-во посетителей в группе A: {}\n'.format(visitors[visitors['group']=='A']['visitors'].sum()))
print('Кол-во покупателей в группе B: {}'.format(len(ordersByUsersB)))
print('Кол-во покупок в группе B: {}'.format(ordersByUsersB['orders'].sum()))
print('Кол-во посетителей в группе B: {}'.format(visitors[visitors['group']=='B']['visitors'].sum()))
```

**Нулевая** гипотеза говорит о том, что **статистическая значимость не значительна и сделать вывод о различии нельзя**, **альтернативная** гипотеза говорит о том, что **статистическая значимость значитаельная и между выборками имеется статистическая значимость**. Порогом статистической значимости установим alpha=.05

```python
#составим списки кол-ва заказов sampleA и sampleB со всеми пользователями (в том числе и не купившими) 
#по группам 
list_orders_1 =[]
for i in range(0,(visitors_filt[visitors_filt['group']=='A']['visitors'].sum()-len(ordersByUsersA))):
    list_orders_1.append(0)
orders_by_non_purchased_users_A = pd.Series(data = list_orders_1, name = 'orders')

list_orders_2 =[]
for i in range(0,(visitors_filt[visitors_filt['group']=='B']['visitors'].sum()-len(ordersByUsersB))):
    list_orders_2.append(0)
orders_by_non_purchased_users_B = pd.Series(data = list_orders_2, name = 'orders')

sampleA = pd.concat([ordersByUsersA['orders'], orders_by_non_purchased_users_A], axis=0)
sampleB = pd.concat([ordersByUsersB['orders'], orders_by_non_purchased_users_B], axis=0)

alpha = 0.05
results = stats.mannwhitneyu(sampleA, sampleB, alternative='two-sided')
print('P-value по конверсии до очистки данных: {}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
print('Отношения средних B/A по конверсии (до очистки данных): {:.3f}'.format(sampleB.mean()/sampleA.mean()))
```

```python
sampleAFiltered = pd.concat([
    ordersByUsersA[np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))]['orders'],
    orders_by_non_purchased_users_A
], axis = 0)
sampleBFiltered = pd.concat([
    ordersByUsersB[np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))]['orders'],
    orders_by_non_purchased_users_B
], axis = 0)

results = stats.mannwhitneyu(sampleAFiltered, sampleBFiltered, alternative='two-sided')
print('P-value по конверсии после очистки данных: {:.4f}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
print('Отношения средних B/A по конверсии (после очистки данных): {:.3f}'.format
                                                    (sampleBFiltered.mean()/sampleAFiltered.mean()))
```
